```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Movimientos_cuenta_0184491.xls',
                 header=0,
                   skiprows=2,
                   names=[ 'Fecha','Fecha valor','Movimiento','Mas datos','Importe','Saldo'])
df.head()
```

```python
df.plot()
```

```python
plt.plot(df['Fecha valor'], df['Importe'], color='r', label='Importe')
plt.plot(df['Fecha valor'], df['Saldo'], color='g', label='Saldo')

plt.xlabel("Fecha") 
plt.ylabel("Dinero") 
plt.title("Importe y Saldo")
plt.legend() 
plt.show() 
##plt.figure(figsize=(32, 32))
```
